# Techinques to Apply on Nominal Categorical Variables 

## 3. One Hot Encoding - On Many Categories 

*  using 10 most frequent labels convert them into dummy variables using onehotencoding

```python
# let's find the top 10 most frequent categories for the variable(column) X2

df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
df = pd.read_csv("data/mercedes.csv", usecols = ['X1','X2','X3','X4','X5','X6'])

# Method to get top 10 most occuring values in a column

# value_counts() --> gets unique values
def get_top10_labels(df,col):
    return [x for x in df[col].value_counts().sort_values(ascending=False).head(10).index]


# Apply one hot encoding for columns and drop them 
def apply_one_hot_coding(df, cols):
    for col in cols:
        col_top10_labels = get_top10_labels(df,col)  
        
        # get whole set of dummy variables (binary variables), for all the categorical variables
        
        for label in col_top10_labels:
            df[col+"_"+label] = np.where(df[col] == label, 1,0) # new columns in dataframe are created here        
        
        df = df.drop(col, axis=1) # Drop current column 
    return df

# if within top 10 --> 1, else --> 0
```

```python
df = apply_one_hot_coding(df, ['X1','X2','X3','X4','X5','X6'])
df.head()
```
