# Techinques to Apply on Nominal Categorical Variables 

## 2. One Hot Encoding - Scikit-Learn library

* [One Hot Encoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html#sklearn.preprocessing.OneHotEncoder)

```python
X = df.drop('price',axis=1)
X
```

```python
y = df['price']
y
```

```python
# Turn the categories (town) into numbers

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["town"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([
    ("one_hot",one_hot,categorical_features)],
    remainder="passthrough")

transformed_X = transformer.fit_transform(X)
transformed_X

# remainder="passthrough" --> ignore all other columns other than the ones mentioned above
```

```python
# all data is now numerical
# The nominal categories are now either 1 or 0

final_df = pd.DataFrame(transformed_X)
final_df.rename(columns={0: "monroe township", 1: "west windsor",2:"robinsville",3:"area"},inplace=True)
final_df
```
