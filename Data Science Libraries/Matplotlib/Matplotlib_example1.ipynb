```python
%matplotlib inline 
# it is a magic command that states import all graphs and plots
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### 3 ways of creating plots

```python
# Create Data
x = np.array([1,2,3,4]) # or list
y = np.array([11,22,33,44])
```

```python
# plot() --> plot x vs y
# plot() function draws a line from point to point.
# Parameter 1 is an array containing the points on the x-axis.
# Parameter 2 is an array containing the points on the y-axis.

# plot() and figure() mean the same thing

# 1st method - OOP
fig = plt.figure() # creates a figure 
ax = fig.add_subplot() # adds some axis
plt.show()
```

```python
# 2nd method - also OOP
# Easier and more robust going forward 

fig, ax = plt.subplots()  # Create a figure and a set of subplots
ax.plot(x, y); # semi-colon needed here 
```

```python
# 3rd method - loose  
plt.plot(x,y)
plt.show()
```

```python
# marker - emphasize each point (dot) 

# ==> "marker:line:color" (for graph)
# marker="o" - ring/"*" - star/"X" - x/"+" - plus, etc
# line -->"-" = solid line/":" = dotted/ "--" = dashed
# color --> "r" = red, "g" = green, "b" = blue, or hexadecimals

# we can also specidy marker size using ms
# mec --> marker edge color, mfc = marked face color
```

```python
# 1) Prepare Data
x = np.array([1,2,3,4]) # or list
y = np.array([11,22,33,44])

# 2) Setup Plot
# figsize(width,height) --> dimensions of the plot
fig, ax = plt.subplots(figsize=(10,10)) # Create a figure and a set of subplots.

# 3) Plot data
# ax.plot(x,y, marker='o', ms=20, mec="g", mfc="g")
ax.plot(x,y, "o-r", ms=20, mec="g", mfc="g")

# 4) Customize data
ax.set(title="Simple plot",
       xlabel="x-axis",
       ylabel="y-axis",
       ylim = (0,50), # set the range of the axis
       xlim = (0,5.0)); # add ; in last

# grid() function to add grid lines to the plot
# axis parameter in the grid() function to specify which grid lines to display - e.g .grid(axis = 'x/y')
ax.grid(color="b",ls="-",lw=2)

# 5) Save the figure and show
fig.savefig("sample-plot.png")

# title() function to set a title for the plot
# xlabel() and ylabel() functions to set a label for the x- and y-axis.
# learn how to change font, position in ax.set -- axis subplot
```

```python
x = np.array([1,2,3,4]) # or list
y = np.array([11,22,33,44])

# To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

fig, ax = plt.subplots(figsize=(10,10))

ax.plot(x,y, "o", ms=20, mec="g", mfc="g")

ax.set(title="Simple plot",
       xlabel="x-axis",
       ylabel="y-axis",
       ylim = (0,50),
       xlim = (0,5.0)); 

ax.grid(color="b",ls="-",lw=2)
```

### Line
Line is the default type of visualization in Matplotlib. 
Usually, unless specified otherwise, your plots will start out as lines.

If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3, (etc. depending on the length of the y-points.

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
fig,ax = plt.subplots() # find out the default fig-size
# plotting two lines
ax.plot(y1,ls = '--',c="r",lw="3");
ax.plot(y2,ls = '--',c="b",lw="3");

# linestyle ---> ls
# ls = ":" = dotted, ls = "--" = dashed

# color --> c

# linewidth --> lw
```

```python
# plotting a curve 

x = np.linspace(0,10,100) # return 100 evenly (diff.) float spaced intervals between 0 and 10
fig, ax = plt.subplots()
ax.plot(x,x**2); # y=x^2
```

### Scatter Plot

```python
# scatter() function to draw a scatter plot

# exponential curves
x = np.linspace(0,10,100)
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Trigonometric curves - sin wave
x = np.linspace(0,10,100)
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

### Bar

* Vertical Bar
* Horizontal Bar

```python
# create a plot from a dictionary
nut_butter_prices = {
    "Almond butter":10,
    "Peanut butter":8,
    "Cashew butter":12
}

# vertical bar
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values(),color = "r",width = 0.2)
ax.set(title="Butter Prices Store",  
       ylabel="Price ($)"); # add ; in last
# default width value is 0.8

# or
# x = np.array(["A", "B", "C", "D"])
# y = np.array([3, 8, 1, 10])
```

```python
# horizontal bar
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()),height = 0.4)
ax.set(title="Butter Prices Store",  
       xlabel="Price ($)");
# for barh --> values must turn into a list
# For horizontal bars, use height instead of width
# The default height value is 0.8s


# or
# x = np.array(["A", "B", "C", "D"])
# y = np.array([3, 8, 1, 10])
```

### Histogram (hist)

* Could show image of normal distribution here
* A histogram is a graph showing frequency distributions.

```python
# Make some data from a normal distribution
x = np.random.randn(1000) # pulls data from a normal distribution curve
# randn = normal distribution
# pulls 1000 random data
# or x = np.random.normal(170, 10, 250) - normal data distribution
fig, ax = plt.subplots()
ax.hist(x);

# np.random.random(1000) --> random data from random distribution
```

## Subplots - Multiple plots on one figure

* https://matplotlib.org/3.1.1/gallery/recipes/create_subplots.html

### 2 ways to create sub plots

```python
# 1st option - recommended 

# subplots() function describes the layout of the figure
# You can draw as many plots you like on one figure, just descibe the number of rows, columns

fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
# a tuple on ax1,ax2,ax3,ax4 --> 4 smaller graphs on a single large graph
# 2 rows, 2 cols on subplot grid

# plot to each different axis
ax1.plot(x,x/2); # axis = 0 
ax2.scatter(np.random.random(10),np.random.random(10));  # axis = 1
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());  # axis = 2
ax4.hist(np.random.randn(1000)); # randn = normal distribution  # axis = 3

# You can add a title to each plot with the title()
# You can add a title to the entire figure with the suptitle()
```

```python
# Option 2: 
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different index - list way
ax[0, 0].plot(x, x/2); # ax[0,0] - axis 0 (row 0, col 0)
ax[0, 1].scatter(np.random.random(10), np.random.random(10)); # ax[0,1] - axis 1 (row 0, col 1)
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Pie Chart

* pie() function to draw pie charts

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

# startangle - default start angle (can be changed)

fig, ax = plt.subplots()
ax.pie(y,labels=mylabels,startangle = 90,colors=mycolors);
ax.legend(title="Some Info");
```
